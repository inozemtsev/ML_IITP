```python
# к первой лабе
import sklearn.datasets
import numpy
import matplotlib.pylab as plt
from sklearn.linear_model import LogisticRegression, Perceptron 
```

```python
## Формирование датасета (X, Y)
iris = sklearn.datasets.load_iris()
X = iris.data[:, :2]
Y = numpy.zeros(len(iris.target))
Y[iris.target>0] = 1

## Обучение персептрона
model = LogisticRegression()
model.fit(X, Y)

# вытаскиваем коэффициенты из модели
a = model.coef_[0][0]
b = model.coef_[0][1]
c = model.intercept_[0]

## Отрисовка результата
iSelector = Y==0
plt.scatter(X[iSelector, 0], X[iSelector, 1], c='b')
plt.scatter(X[~iSelector, 0], X[~iSelector, 1], c='y')
# рисуем по ним прямую
plt.plot([4, 8], [(-a*4 - c)/b , (-a*8 - c)/b])
plt.show()
```

# # Переобучение (overfitting)

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

n_samples = 12
# настоящая функция
true_fun = lambda X: np.cos(1.5 * np.pi * X)
# зашумленная выборка
X = np.sort(np.random.rand(n_samples))
X = X[:, np.newaxis]

y = true_fun(X[:, 0]) + np.random.randn(n_samples) * 0.1
Axe = np.linspace(0, 1,100)[:, np.newaxis]
# отрисовка
plot(Axe, true_fun(Axe[:,0]))
scatter(X[:, 0], y)
```

```python
degree = 7
# переводим в расширенное признаковое пространство 
transf = PolynomialFeatures(degree, include_bias=False)
X_train = transf.fit_transform(X)
X_test = transf.fit_transform(Axe)
# находим разделяющую гиперплоскость
model = LinearRegression()
model.fit(X_train, y)
# отрисовываем
figsize(7,5)
plot(Axe, model.predict(X_test), label='Model')
plot(Axe, true_fun(Axe[:,0]), label='True')
ylim(-1.5, 1.5)
scatter(X[:,0], y)
```

```python
# соберем контрольное множество
n_samples = 10
true_fun = lambda X: np.cos(1.5 * np.pi * X)

X = np.sort(np.random.rand(n_samples))
X = X[:, np.newaxis]
y = true_fun(X[:, 0]) + np.random.randn(n_samples) * 0.1

X_holdout = np.sort(np.random.rand(n_samples*2))
X_holdout = X_holdout[:, np.newaxis]
y_holdout = true_fun(X_holdout[:, 0])

degree_list = range(1,15)#[1, 2, 3, 5, 7, 10]
err_train = []
err_holdout = []
for degree in degree_list:
    # переводим в расширенное признаковое пространство 
    transf = PolynomialFeatures(degree, include_bias=False)
    X_train = transf.fit_transform(X)
    X_test = transf.fit_transform(X_holdout)
    # находим разделяющую гиперплоскость
    model = LinearRegression()
    model.fit(X_train, y)
    # считаем ошибки
    err_train.append(log(mean((model.predict(X_train) - y)**2)))
    err_holdout.append(log(mean((model.predict(X_test) - y_holdout)**2)))
plot(degree_list, err_train)
plot(degree_list, err_holdout)
```

# # Регуляризация

```python
## Формирование датасета (X, Y)
iris = sklearn.datasets.load_iris()
X = iris.data[:, :2]
Y = numpy.zeros(len(iris.target))
Y[iris.target>0] = 1

## Обучение персептрона
model = LogisticRegression()
model.fit(X, Y)

# вытаскиваем коэффициенты из модели
a = model.coef_[0][0]
b = model.coef_[0][1]
c = model.intercept_[0]

def draw_plot(w):
    ## Отрисовка результата
    iSelector = Y==0
    plt.scatter(X[iSelector, 0], X[iSelector, 1], c='b')
    plt.scatter(X[~iSelector, 0], X[~iSelector, 1], c='y')
    # рисуем по ним прямую

    plt.plot([4, 8], [(-w[0]*4 - w[2])/w[1] , (-w[0]*8 - w[2])/w[1]])
    plt.show()

draw_plot(np.array([a, b, c]))
```

```python
from scipy.optimize import basinhopping
def calc_loss(w, C):
    y_pred = ((np.dot(X, w[:2]) + w[2]) > 0)*1
    return np.linalg.norm(y_pred - Y) + C*np.linalg.norm(w, ord=2)
```

```python
target_func = lambda x: calc_loss(np.array([1, x[0], x[1]]), 0.01)
opt = basinhopping(target_func, x0=np.array([2.0,3.0]), stepsize=0.5, niter=2000)
w = array([1, opt['x'][0], opt['x'][1]])
print opt
draw_plot(w)
```

```python
from sklearn import decomposition
from sklearn.datasets import fetch_mldata
import pickle
import logging
reload(logging)

logging.basicConfig(format = u'[%(asctime)s]  %(message)s', level = logging.INFO)
logging.info("Test output")
def download_mnist(path="./data/mnist/"):
    #Скачивает базу MNIST (~55мб)
    mnist = fetch_mldata('MNIST original', data_home=path)
    pickle.dump(mnist, open("./data/mnist/mnist.pickled", "w"), -1)
    
def load_mnist():
    #Загружает базу MNIST из дампа и уменьшает размерность до 20 (PCA)
    mnist = pickle.load(open("./data/mnist/mnist.pickled", "r"))
    pca = decomposition.PCA(n_components=40)
    logging.info("Start PCA...")
    pca.fit(mnist['data'][1:20000,:])
    X_raw = pca.transform(mnist['data'])#[1:40000,:])
    logging.info("Done!")
    return (X_raw, mnist['target'], 'mnist')
```
