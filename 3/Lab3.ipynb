```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

```python
# Одно решающее дерево
```

```python
# Генерируем точки
rng = np.random.RandomState(2014)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))

# Функция визуализации
def draw_one_tree_prediction(clf, X, Y, col='g'):
    X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
    y_pred = clf.predict(X_test)

    scatter(X, Y, c="k", label="data")
    plot(X_test, y_pred, c=col, linewidth=2)
    xlabel("data")
    ylabel("target")
```

```python
# Decision Tree
model = DecisionTreeRegressor(max_depth=2)
model.fit(X, y)

draw_one_tree_prediction(model, X, y)
```

```python
# Отрисовка дерева 
from sklearn.tree import export_graphviz
from IPython.core.display import Image
export_graphviz(model, out_file='tree.dot')
# следующая строчка отработает только на Linux c установленным пакетом graph-viz
! dot -Tpng tree.dot -o tree.png
Image(filename='tree.png') 
```

```python
# Случайный лес

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=2, max_depth=2, bootstrap=True)
model.fit(X, y)

# Красное и зеленое - результат отдельных деревьев
draw_one_tree_prediction(model.estimators_[0], X, y, col='r')
draw_one_tree_prediction(model.estimators_[1], X, y, col='g')
# Желтое - среднее по лесу
draw_one_tree_prediction(model, X, y, col='y')
```

```python
# Бустинг над решающими деревьями

from sklearn.ensemble import AdaBoostRegressor

# Генерация точек
rng = np.random.RandomState(1)
X = np.linspace(0, 6, 100)[:, np.newaxis]
y = np.sin(X).ravel() + np.sin(6 * X).ravel() + rng.normal(0, 0.1, X.shape[0])

# Построение бустинговой композиции из трех деревьев
clf_boost = AdaBoostRegressor(DecisionTreeRegressor(max_depth=3),
                          n_estimators=3, random_state=rng, loss='square', learning_rate=2)
clf_boost.fit(X, y)
y_boost = clf_boost.predict(X)

# Отрисовка результата
draw_one_tree_prediction(clf_boost.estimators_[0],  X, y,  col='r')
draw_one_tree_prediction(clf_boost.estimators_[1],  X, y,  col='g')
draw_one_tree_prediction(clf_boost.estimators_[2],  X, y,  col='b')
```
